
30. Write a Pandas program to create a side-by-side bar plot of the diamonds
DataFrame.


```python
df_1 = df.groupby('cut')['price'].mean()
df_2 = df.groupby('color')['price'].mean()
df_3 = df.groupby('clarity')['price'].mean()

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.barplot(x=df_1.index, y=df_1.values, palette="viridis")
plt.title('Average Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Average Price')

plt.subplot(1, 3, 2)
sns.barplot(x=df_2.index, y=df_2.values, palette="magma")
plt.title('Average Price by Color')
plt.xlabel('Color')
plt.ylabel('Average Price')

plt.subplot(1, 3, 3)
sns.barplot(x=df_3.index, y=df_3.values, palette="cool")
plt.title('Average Price by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Average Price')

plt.tight_layout()
plt.show()
```
