12. Write a Python program to create bar plots with error bars on the same
figure.
Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
x_values = [1, 2, 3, 4, 5]
avg_velocity = [0.2474, 0.1235, 0.1737, 0.1824, 0.1025]
std_dev = [0.3314, 0.2278, 0.2836, 0.2645, 0.2250]

error_bars = [val / 8 for val in std_dev]

plt.bar(x=x_values, height=avg_velocity, zorder=2)

plt.errorbar(x_values, avg_velocity, yerr=error_bars, fmt='o', color='black', zorder=2, capsize=2)

plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('Bar Plot with Error Bars')
plt.grid(which='major', color='blue', linestyle='-', linewidth=0.5, zorder=1)
plt.minorticks_on()
plt.grid(which='minor', color='gray', linestyle='--', linewidth=0.2, zorder=1)
plt.show()
```

13. Write a Python program to create bar plots with errorbars on the same
figure. Attach a text label above each bar displaying men means (integer
value).
Sample Data
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
x_values = [1, 2, 3, 4, 5]
avg_velocity = [0.2474, 0.1235, 0.1737, 0.1824, 0.1025]
std_dev = [0.3314, 0.2278, 0.2836, 0.2645, 0.2250]

avg_velocity = [val * 18/5 for val in avg_velocity]
std_dev = [val * 18/5 for val in std_dev]

error_bars = [val / 9 for val in std_dev]

plt.figure(figsize=(8,7))
plt.bar(x=x_values, height=avg_velocity, yerr=error_bars, capsize=5, zorder=2)

for i in range(len(avg_velocity)):
    plt.text(
        x=x_values[i],
        y=avg_velocity[i] + 0.16,
        s=f"{avg_velocity[i]:.4f}",
        ha='center',
        bbox=dict(facecolor='white', boxstyle='round,pad=0.3')  # No edgecolor specified
    )


y_min, y_max = 0, max(avg_velocity) + 0.3
plt.yticks(np.linspace(y_min, y_max, 6))

plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('Bar Plot with Error Bars')
plt.grid(which='major', color='blue', linestyle='-', linewidth=0.5, zorder=1)
plt.minorticks_on()
plt.grid(which='minor', color='gray', linestyle='--', linewidth=0.2, zorder=1)
plt.show()
```

14. Write a Python program to create a stacked bar plot with error bars.
Note: Use bottom to stack the women?s bars on top of the men?s bars.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
Men Standard deviation = (4, 3, 4, 1, 5)
Women Standard deviation = (3, 5, 2, 3, 3)

```python
mean_men = [22, 30, 35, 35, 26]
mean_women = [25, 32, 30, 35, 29]
men_std_dev = [4, 3, 4, 1, 5]
women_std_dev = [3, 5, 2, 3, 3]

x = np.arange(len(mean_men))
width = 0.5

fig, ax = plt.subplots(figsize=(10, 5))

bar_men = ax.bar(x, mean_men, width, label='Men', color='lightblue',yerr=men_std_dev, zorder=2, capsize=1.5)
bar_women = ax.bar(x, mean_women, width, bottom=mean_men, label='women', color='pink',yerr=women_std_dev, zorder=2, capsize=1.5)

ax.set_ylabel('Scores')
ax.set_title('Stacked Bar Chart with Error Bars')
ax.set_xticks(x)
ax.set_xticklabels(['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5'])
ax.legend()
plt.show()
```

15. Write a Python program to create a horizontal bar chart with differently
ordered colors.
Note: Use bottom to stack the women?s bars on top of the men?s bars.
Sample Data Set:
languages = [['Language','Science','Math'],
['Science','Math','Language'],
['Math','Language','Science']]
numbers = [{'Language':75, 'Science':88, 'Math':96},
{'Language':71, 'Science':95, 'Math':92},
{'Language':75, 'Science':90, 'Math':89}]

```python
data = {'Math': [85, 90, 78],
        'Science': [88, 92, 80],
        'English': [75, 85, 95]}
df = pd.DataFrame(data, index=['Student 1', 'Student 2', 'Student 3'])

students = list(df.index)
subjects = list(df.columns)

math_scores = df['Math'].tolist()
science_scores = df['Science'].tolist()
english_scores = df['English'].tolist()

fig, ax = plt.subplots(figsize=(10, 6))

bars_math = ax.barh(students, math_scores, color='blue', label='Math')
bars_science = ax.barh(students, science_scores, left=math_scores, color='red', label='Science')
bars_english = ax.barh(students, english_scores, left=np.array(math_scores) + np.array(science_scores), color='green', label='English')

ax.set_xlabel('Scores')
ax.set_title('Student Scores by Subject')
ax.legend()
plt.show()
```

16. Write a Python program to create stack bar plot and add label to each
section.
Sample data:
people = ('G1','G2','G3','G4','G5','G6','G7','G8')
segments = 4
 multi-dimensional data
data = [[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039,
7.1123587, 12.77792868, 3.44773477],
[ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195,
6.79685302, 7.24578743, 3.69371847],
[ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182,
4.63832778, 11.16849999, 8.56883433],
[ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185,
10.97567589, 3.98287652, 8.80552122]]

```python
data_dict = {
    'G1': [3.40022085, 11.24811149, 3.94253354, 4.24409799],
    'G2': [7.70632498, 5.03778215, 4.74763549, 12.71746612],
    'G3': [6.4097905, 6.65808464, 11.73529246, 11.3772169],
    'G4': [10.51648577, 12.32220677, 4.6465543, 9.00514257],
    'G5': [7.5330039, 7.45964195, 12.9952182, 10.47084185],
    'G6': [7.1123587, 6.79685302, 4.63832778, 10.97567589],
    'G7': [12.77792868, 7.24578743, 11.16849999, 3.98287652],
    'G8': [3.44773477, 3.69371847, 8.56883433, 8.80552122]
}

df = pd.DataFrame(data_dict)
df['Segments'] = ['Segment 1', 'Segment 2', 'Segment 3', 'Segment 4']
df = df.set_index('Segments')

fig, ax = plt.subplots(figsize=(10, 6))

left_values = np.zeros(len(df.columns))

for segment in df.index:
    values = df.loc[segment].values
    total = df.sum(axis=0).values
    bars = ax.barh(df.columns, values, left=left_values, label=segment, edgecolor='black', linewidth=1.5)

    for i, value in enumerate(values):
        percentage = (value / total[i]) * 100
        ax.text(left_values[i] + value / 2, i, f"{percentage:.1f}%",
                ha='center', va='center', color='white', fontsize=10)

    left_values += values

ax.set_xlabel('Values')
ax.set_title('Horizontal Stacked Bar Chart: Groups vs Segments with Borders and Percentages')
ax.legend(title='Segments', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

17. Write a Python program to add textures (black and white) to bars and
wedges.
Note: Use bottom to stack the womens bars on top of the mens bars.

```python
fig, ax = plt.subplots(figsize=(10, 6))

patterns = ['|', '/', '//', '+', '-', '.', 'o', 'x', 'O', '*']

for i, pattern in enumerate(patterns):
    ax.add_patch(plt.Rectangle((i, 0), 1, 3,
                               hatch=pattern,
                               fill=False,
                               edgecolor='black'))

ax.set_xlim(0, 10)
ax.set_ylim(0, 3)
ax.grid(False)
ax.set_xlabel('')
ax.set_ylabel('')
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.set_xticks(range(11))
ax.set_yticks([0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0])
plt.tight_layout()
plt.show()
```
